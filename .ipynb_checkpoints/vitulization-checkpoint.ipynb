```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import fetch_california_housing
import seaborn as sns
```

```python
dataset = fetch_california_housing(as_frame = True)

x = pd.DataFrame(dataset.data, columns=dataset.feature_names) 
y = pd.Series(dataset.House_Price, name='House_Price') 


print(x.info)
print(y.info)

print(x.head())
print(y.head())

print(x.describe())
print(y.describe())
```

## 1) Box Plot

```python
fig, axes = plt.subplots(2, 4, figsize = (15, 6))
axes = axes.flatten()
print(x.columns)

for i,name in enumerate(x.columns):
    print(i,name)
    axes[i].boxplot(x[name])
    axes[i].set_xlabel(name)    

plt.show()
```

## 2) Violin Plot

```python
fig, axes = plt.subplots(2, 4, figsize=(15, 6))
axes = axes.flatten()

for i, name in enumerate(x.columns):
    print(i, name)
    sns.violinplot(x=x[name], ax=axes[i])
    axes[i].set_xlabel(name) 

plt.tight_layout()
plt.show()
```

## 3) Crosstabs with pie chart

```python
y_category = pd.cut(y, bins=[0, 1.5, 3, 4.5, np.inf], labels=['Low', 'Medium', 'High', 'Very High'])

x['IncomeGroup'] = pd.cut(x['MedInc'], bins=[0, 2, 4, 6, 8, np.inf], labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
x['HouseAgeGroup'] = pd.cut(x['HouseAge'], bins=[0, 10, 20, 30, 40, 50, np.inf], labels=['0-10', '11-20', '21-30', '31-40', '41-50', '51+'])
x['AveRoomsGroup'] = pd.cut(x['AveRooms'], bins=[0, 2, 4, 6, 8, np.inf], labels=['Tiny', 'Small', 'Medium', 'Large', 'Very Large'])
x['AveBedrmsGroup'] = pd.cut(x['AveBedrms'], bins=[0, 1, 2, 3, np.inf], labels=['Single', 'Small', 'Medium', 'Large'])
x['PopulationGroup'] = pd.cut(x['Population'], bins=[0, 1000, 5000, 10000, np.inf], labels=['Low', 'Medium', 'High', 'Very High'])
x['AveOccupGroup'] = pd.cut(x['AveOccup'], bins=[0, 2, 4, 6, np.inf], labels=['Small', 'Medium', 'Large', 'Very Large'])


df = x.copy()
df['House_PriceCategory'] = y_category  # Add categorized House_Price column


crosstabs = {
    'Income vs House_Price': pd.crosstab(df['House_PriceCategory'], df['IncomeGroup']),
    'House Age vs House_Price': pd.crosstab(df['House_PriceCategory'], df['HouseAgeGroup']),
    'Rooms vs House_Price': pd.crosstab(df['House_PriceCategory'], df['AveRoomsGroup']),
    'Bedrooms vs House_Price': pd.crosstab(df['House_PriceCategory'], df['AveBedrmsGroup']),
    'Population vs House_Price': pd.crosstab(df['House_PriceCategory'], df['PopulationGroup']),
    'Occupancy vs House_Price': pd.crosstab(df['House_PriceCategory'], df['AveOccupGroup'])
}

# Plot Pie Charts for Each Crosstab
fig, axes = plt.subplots(len(crosstabs), len(crosstabs['Income vs House_Price'].index), figsize=(18, 18))

for row, (title, crosstab) in enumerate(crosstabs.items()):
    axes[row] = axes[row].flatten()  # Ensure 1D indexing
    
    for i, House_Price_category in enumerate(crosstab.index):
        axes[row][i].pie(crosstab.loc[House_Price_category], labels=crosstab.columns, autopct='%1.1f%%', startangle=90)
        axes[row][i].set_title(f'{title} ({House_Price_category})')


plt.tight_layout()
plt.show()
```

## 4) Cylindrical Bar Graph

```python
df['Target'] = y 


df['IncomeGroup'] = pd.cut(df['MedInc'], bins=[0, 2, 4, 6, 8, np.inf], labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])


income_group_avg = df.groupby('IncomeGroup')['Target'].mean()


fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')


x_labels = income_group_avg.index  
x_pos = np.arange(len(x_labels)) 
y_pos = np.zeros(len(x_labels))  
z_pos = np.zeros(len(x_labels))  

# Bar dimensions
dx = np.ones(len(x_labels)) * 0.6  
dy = np.ones(len(x_labels)) * 0.6  
dz = income_group_avg.values  

# Plot cylindrical bars
ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, shade=True)

# Label axes
ax.set_xlabel("Income Group")
ax.set_ylabel("Y Axis (Dummy)")
ax.set_zlabel("Average House Price ($100,000s)")
ax.set_title("Cylindrical Bar Graph: Income Group vs. House Price")


ax.set_xticks(x_pos)
ax.set_xticklabels(x_labels)

plt.show()
```
